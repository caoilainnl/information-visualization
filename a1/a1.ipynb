```python
import pandas as pd  # type: ignore
import seaborn as sns  # type: ignore
import matplotlib.pyplot as plt  # type: ignore
from sklearn.datasets import load_iris  # type: ignore
```

```python
"""Load data and create visualizations for the Iris dataset."""
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["species"] = pd.Categorical.from_codes(iris.target, iris.target_names)

sns.set(style="whitegrid", context="notebook")

plt.figure(figsize=(8, 6))
scatter_ax = sns.scatterplot(
    data=df,
    x="petal length (cm)",
    y="petal width (cm)",
    hue="species",
    palette="Set2",
    edgecolor="black",
    alpha=0.8,
)
scatter_ax.set_title("Iris Petal Dimensions by Species")
scatter_ax.set_xlabel("Petal Length (cm)")
scatter_ax.set_ylabel("Petal Width (cm)")
plt.tight_layout()
plt.show()



```

```python
pair_plot = sns.pairplot(df, hue="species", diag_kind="kde", palette="Set2")
pair_plot.fig.suptitle("Iris Pair Plot", y=1.02)
plt.show(pair_plot)
```

```python
mean_sepal = df.groupby("species")["sepal length (cm)"].mean().reset_index()

plt.figure(figsize=(6, 5))
bar_ax = sns.barplot(
    data=mean_sepal,
    hue="species",
    legend=False,
    y="sepal length (cm)",
    palette="Set2",
    edgecolor="black",
)
bar_ax.set_title("Mean Sepal Length by Species")
bar_ax.set_xlabel("Species")
bar_ax.set_ylabel("Sepal Length (cm)")
plt.tight_layout()
plt.show("iris_sepal_bar.png")
```
